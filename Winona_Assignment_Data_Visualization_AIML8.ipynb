# Dataset

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
````

````python
#masukkan csv
df = pd.read_csv('Telco_customer_churn.csv')
````

````python
#cek data
df.head()
````

````python
#cek data
df.info()
````

#Data Cleaning

````python
#cek berapa banyak missing value
missing_values = df.isnull().sum()
missing_values = missing_values[missing_values > 0]
missing_pct = (missing_values / len(df)) * 100
missing_data = pd.DataFrame(
    {
        'Missing Values' : missing_values,
        'Percentage' : missing_pct
    }
)
missing_data
````

````python
#rename missing value menjadi no churn reason
df['Churn Reason'] = df['Churn Reason'].fillna('No Churn Reason')
````

````python
#cek data kosong ' '
sum_blank = (df['Total Charges'] == ' ').sum()
print("Jumlah data blank pada kolom Total Charges =", sum_blank)

#ubah blank menjadi nan
df['Total Charges'] = df['Total Charges'].replace(' ', np.nan)
````

````python
#cek berapa banyak data duplicate
duplicate_rows = df[df.duplicated()]
print(f"Jumlah baris yang terduplikat = {len(duplicate_rows)}")
````

````python
#cek outlier
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
churn_plots = len(numeric_cols)

num_cols = 3
num_rows = (churn_plots + num_cols - 1) // num_cols

fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 5 * num_rows))
axes = axes.flatten()

for i, feature in enumerate(numeric_cols):
    sns.boxplot(x=df[feature], ax=axes[i])
    axes[i].set_title(f'Boxplot of {feature}')

plt.tight_layout()
plt.show()
````

````python
#karena data sudah aman maka lanjut ke step selanjutnya yaitu data visualization
````

# Data Visualization

### 1. Membuat perbedaan persentase dari customer yang churn dan customer yang tidak churn menggunakan pie chart

````python
#menghitung jumlah customer churn dan tidak churn
churn_counts = df['Churn Value'].value_counts()
labels = ['Tidak Churn', 'Churn']
colors = ['#23C552', '#F84F31']

#membuat pie chart
plt.pie(churn_counts, labels=labels, colors=colors, autopct='%1.1f%%')
plt.title('Customer Churn VS Tidak Churn')
plt.show()
````

### 2. Melihat hubungan antara `jumlah bulan yang telah dihabiskan customer`  dan `total tagihan customer` menggunakan scatter plot

````python
#ubah kolom Total Charges menjadi numeric
df['Total Charges'] = pd.to_numeric(df['Total Charges'])

#nilai nan diganti menjadi 0
df['Total Charges'] = df['Total Charges'].fillna(0)
````

````python
#membuat scatter plot
plt.figure(figsize=(10, 6))
scatter_plot = sns.regplot(data=df, x='Tenure Months', y='Total Charges')
plt.title('Hubungan antara Lama Berlangganan dengan Total Tagihan')
plt.xlabel('Lama Berlangganan (dalam Bulan)')
plt.ylabel('Total Tagihan')
plt.show()
````

### 3. Melihat 5 kota teratas yang customernya melakukan churn dan tidak melakukan churn menggunakan multi bar chart

````python
#mengambil total churn label di setiap kota
city_churn = df.groupby(['City', 'Churn Label']).size().unstack(fill_value=0)

#mengambil top 5
top5_cities = city_churn.sum(axis=1).nlargest(5).index
top5_cities_churn = city_churn.loc[top5_cities]
print(top5_cities_churn)
````

````python
#mempersiapkan data untuk visualisasi
top5 = top5_cities_churn.reset_index().melt(
    id_vars = 'City',
    var_name = 'Churn Label',
    value_name = 'Jumlah'
)

#membuat multi bar chart
plt.figure(figsize=(10, 6))
sns.barplot(data=top5, x='City', y='Jumlah', hue='Churn Label', palette=['#23C552', '#F84F31'])
plt.title('Perbandingan 5 Kota Teratas Berdasarkan Status Churn')
plt.xlabel('Kota')
plt.ylabel('Jumlah Customer')
plt.tight_layout()
plt.show()
````

### 4. Melihat persebaran/distirbusi data dari `tenure months` untuk masing-masing `churn label` memakai boxplot

````python
#membuat boxplot
sns.boxplot(data=df, x='Churn Label', y='Tenure Months', hue='Churn Label', palette=['#F84F31', '#23C552'])
plt.title('Distribusi Data Lama Berlangganan Berdasarkan Status Churn')
plt.xlabel('Status Churn')
plt.ylabel('Lama Berlangganan (dalam Bulan)')
plt.tight_layout()
plt.show()
````
